```python
NETWORK_LIMIT_PER_NODE = 200 / 8
PMEM_READ_LIMIT_PER_NODE = 80.309346 * 2**30 / 10**9
PMEM_WRITE_LIMIT_PER_NODE = 25.836094 * 2**30 / 10**9

print(f"PMEM_READ_LIMIT_PER_NODE: {PMEM_READ_LIMIT_PER_NODE} GB/s")
print(f"PMEM_WRITE_LIMIT_PER_NODE: {PMEM_WRITE_LIMIT_PER_NODE} GB/s")
print(f"NETWORK_LIMIT_PER_NODE: {NETWORK_LIMIT_PER_NODE} GB/s")

def network_limit(nnodes):
    return nnodes * NETWORK_LIMIT_PER_NODE

def pmem_read_limit(nnodes):
    return nnodes * PMEM_READ_LIMIT_PER_NODE

def pmem_write_limit(nnodes):
    return nnodes * PMEM_WRITE_LIMIT_PER_NODE
```

```python
re_tile_io = {
    'node': re.compile(r'# mpi-tile-io run on (\w+)'),
    'processes': re.compile(r'# (\d+) process\(es\) available, (\d+) used'),
    'filename': re.compile(r'# filename: (.+)'),
    'collective_io': re.compile(r'# collective I/O (on|off)'),
    'header_size': re.compile(r'# (\d+) byte header'),
    'dataset_info': re.compile(r'# (\d+) x (\d+) element dataset, (\d+) bytes per element'),
    'tiles': re.compile(r'# (\d+) x (\d+) tiles, each tile is (\d+) x (\d+) elements'),
    'tile_overlap': re.compile(r'# tiles overlap by (\d+) elements in X, (\d+) elements in Y'),
    'file_size': re.compile(r'# total file size is ~([\d.]+) Mbytes, (\d+) file\(s\) total\.'),
    'times': re.compile(r'# (Open|Read|Close): min_t = ([\d.]+), max_t = ([\d.]+), mean_t = ([\d.]+), var_t = ([\d.]+)'),
    'read_bandwidth': re.compile(r'Read Bandwidth = ([\d.]+) Mbytes/sec')
}

def load_tile_io_stdout(tile_file: Path):
    try:
        content = tile_file.read_text()

        data = { 
            'node': re_tile_io['node'].search(content).group(1),
            'processes_available': int(re_tile_io['processes'].search(content).group(1)),
            'processes_used': int(re_tile_io['processes'].search(content).group(2)),
            # 'filename': re_tile_io['filename'].search(content).group(1),
            'collective_io': re_tile_io['collective_io'].search(content).group(1) == 'on',
            'header_size': int(re_tile_io['header_size'].search(content).group(1)),
            'dataset_x': int(re_tile_io['dataset_info'].search(content).group(1)),
            'dataset_y': int(re_tile_io['dataset_info'].search(content).group(2)),
            'bytes_per_element': int(re_tile_io['dataset_info'].search(content).group(3)),
            'tiles_x': int(re_tile_io['tiles'].search(content).group(1)),
            'tiles_y': int(re_tile_io['tiles'].search(content).group(2)),
            'tile_elements_x': int(re_tile_io['tiles'].search(content).group(3)),
            'tile_elements_y': int(re_tile_io['tiles'].search(content).group(4)),
            'tile_overlap_x': int(re_tile_io['tile_overlap'].search(content).group(1)),
            'tile_overlap_y': int(re_tile_io['tile_overlap'].search(content).group(2)),
            'total_file_size_mbytes': float(re_tile_io['file_size'].search(content).group(1)),
            'total_files': int(re_tile_io['file_size'].search(content).group(2)),
            'read_bw': float(re_tile_io['read_bandwidth'].search(content).group(1)) * 2**20
        }

        for op_type, min_t, max_t, mean_t, var_t in re_tile_io['times'].findall(content):
            data[f'{op_type.lower()}_min_t'] = float(min_t)
            data[f'{op_type.lower()}_max_t'] = float(max_t)
            data[f'{op_type.lower()}_mean_t'] = float(mean_t)
            data[f'{op_type.lower()}_var_t'] = float(var_t)

        return [data]

    except Exception as e:
        print(f"Failed to load {tile_file}, {e}")
        return []


# load_tile_io_stdout(Path("/work/NBB/hiraga/work/mpiio-pmembb/raw/ior-tile-pmembb/2024.10.03-14.33.03-default/2024.10.03-14.33.49-399118.nqsv-16/tile_stdout_1.txt"))
```

```python
RAW_BASEDIR = Path(f"../../raw/")

default_metadata = {}

# fmt: off
job_group = (
    ("ior-tile-pmembb/2024.10.03-14.33.03-default", {}), # 16 +aggread
    ("ior-tile-pmembb/2024.10.03-14.47.32-default", {}), # 16
    ("ior-tile-pmembb/2024.10.03-14.47.47-default", {}), # 
    ("ior-tile-pmembb/2024.10.03-15.06.11-default", {}), # +aggread
    # ("ior-tile-pmembb/2024.10.03-16.27.19-16kx512", {}),
)
# fmt: on

re_timestamp = re.compile(
    r"([0-9]{4})\.([0-9]{2})\.([0-9]{2})-([0-9]{2})\.([0-9]{2})\.([0-9]{2})"
)


def parse_job_start_timestamp(job_dir: Path):
    m = re_timestamp.search(str(job_dir))
    if m:
        return m.group()
    else:
        return None


re_runid = re.compile(r"_([0-9]+)\.txt")


def get_runid(file: Path):
    m = re_runid.search(str(file))
    if m:
        return int(m.group(1))
    else:
        return None


def load_job_params(job_dir: Path, runid: int):
    try:
        return json.loads((job_dir / f"job_params_{runid}.json").read_text())
    except Exception as e:
        print(f"Failed to load job_params{runid}.json, {e}")
        return {}


df_src = (
    pl.DataFrame(
        [
            {
                **default_metadata,
                "job_start": parse_job_start_timestamp(job_dir),
                **load_job_params(job_dir, get_runid(tile_file)),
                **tile_result,
                **metadata,
            }
            for job_group_dir, metadata in job_group
            for job_dir in (RAW_BASEDIR / job_group_dir).glob("*")
            if job_dir.is_dir()
            for tile_file in job_dir.glob("tile_stdout_*.txt")
            for tile_result in load_tile_io_stdout(tile_file)
        ]
    )
    # .with_columns(
    #     odirect=pl.col("Command line").str.contains("--posix.odirect"),
    #     strided=pl.col("summary_transferSize") == pl.col("summary_blockSize"),
    # )
    # .rename(
    #     {
    #         "summary_bwMaxMIB": "bw",
    #         "summary_MeanTime": "time",
    #         "summary_segmentCount": "segmentCount",
    #         "summary_blockSize": "blockSize",
    #         "summary_transferSize": "transferSize",
    #         "summary_operation": "operation",
    #         "Parameters_filePerProc": "fpp",
    #         "Parameters_reorderTasks": "reorder",
    #         "summary_API": "api",
    #         "result_access": "access",
    #         "result_bwMiB": "bwMiB",
    #         "result_openTime": "openTime",
    #         "result_wrRdTime": "wrRdTime",
    #         "result_closeTime": "closeTime",
    #         "result_totalTime": "totalTime",
    #     }
    # )
    # .with_columns(
    #     fpp=pl.col("fpp").cast(pl.Boolean),
    #     reorder=pl.col("reorder").cast(pl.Boolean),
    #     bwGB=pl.col("bwMiB") * 2**20 / 10**9,
    # )
    .with_columns(
        pl.when((pl.col("storageSystem") == "PEANUTS") & (pl.col("spack_env_name") == "mpiio-pmembb"))
        .then(True)
        .otherwise(False)
        .alias("aggread"),
        (pl.col("read_bw") / 10**9).alias("bwGB"),
        (pl.col("nnodes") // pl.col("tiles_y") - 1).alias("remote_nodes"),
    )
)
```

```python
group_by_columns = [
    "storageSystem",
    "aggread",
    "nnodes",
    "remote_nodes",
]

df = (
    df_src.lazy()
    .group_by(group_by_columns)
    .agg(pl.max("read_min_t"), pl.max("read_max_t"))
    .with_columns(
        pl.when((pl.col("storageSystem") != "PEANUTS"))
        .then(pl.col("storageSystem"))
        .when(pl.col("aggread"))
        .then(pl.lit("PEANUTS w/ aggread"))
        .otherwise(pl.lit("PEANUTS w/o aggread"))
        .alias("legend"),
        (pl.col("read_max_t") - pl.col("read_min_t")).alias("max-min"),
    )
    .sort("nnodes", "remote_nodes")
    .collect()
)


for (legend,),df_sub in df.group_by("legend"):
    for target in ("read_min_t", "read_max_t", "max-min"):
        fig, axes = plt.subplots(ncols=1, figsize=(8, 6), sharey=True)
        fig.patch.set_alpha(1)
        print(legend, target)
        # display(df_sub)
        (
            df_sub
            # .filter(pl.col("remote_nodes") > 0)
            .pivot(values=target, index="nnodes", on="remote_nodes")
            .pipe(lambda df: (display(df), df)[-1])
            .to_pandas()
            .set_index("nnodes")
            .plot(
                ax=axes,
                kind="bar",
                xlabel="Number of nodes",
                ylabel=target,
                rot=0,
                width=0.75,
            )
        )
        axes.grid(which="major", axis="y", alpha=0.8, linestyle="--", linewidth=1)
        plt.show()
```
